```python
# Check exact full-row duplicates first (these ARE safe to drop)
print(df.duplicated().sum())

# Now check duplicates on a business key (likely: train id + station + planned time)
# adjust column names to match your actual dataset
key_cols = ['ID', 'station', 'arrival_plan']  # example
dupe_mask = df.duplicated(subset=key_cols, keep=False)
print(dupe_mask.sum())

# Look at a few examples side by side
df[dupe_mask].sort_values(key_cols).head(20)
```

```python

df[['arrival_delay_check', 'departure_delay_check']].value_counts()
```

```python
# Recompute delay from raw times where possible, rather than trusting delay_m blindly
df['arrival_delay_m_calc'] = (df['arrival_change'] - df['arrival_plan']).dt.total_seconds() / 60
# Compare to existing column to sanity check
(df['arrival_delay_m'] - df['arrival_delay_m_calc']).abs().describe()
```

```python
# Inspect what values exist
df['info'].value_counts(dropna=False)
```

```python
#Exploratory Analysis
```

```python
#overall punctuality baseline
df['is_delayed'] = df['arrival_delay_m'] > 6  # DB's own official punctuality threshold is 6 min
df['is_delayed'].mean()  # overall % delayed
```

```python
#worst station
station_stats = df.groupby('station').agg(
    avg_delay=('arrival_delay_m', 'mean'),
    pct_delayed=('is_delayed', 'mean'),
    n_stops=('ID', 'count')
).query('n_stops >= 20').sort_values('avg_delay', ascending=False)  # filter out low-traffic noise
station_stats.head(20)
```

```python
# wost line / train types
df['train_type'] = df['line'].str.extract(r'^([A-Za-z]+)')  # e.g., ICE, RE, RB
df.groupby('train_type')['arrival_delay_m'].mean().sort_values(ascending=False)
```

```python
#when delays spike - time paterns
df['hour'] = df['arrival_plan'].dt.hour
df['weekday'] = df['arrival_plan'].dt.day_name()

df.groupby('hour')['arrival_delay_m'].mean().plot(kind='line', title='Avg delay by hour')
df.groupby('weekday')['arrival_delay_m'].mean().reindex(
    ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
).plot(kind='bar', title='Avg delay by weekday')
```

```python
# Heat map hour vs weekday
import seaborn as sns
pivot = df.pivot_table(values='arrival_delay_m', index='weekday', columns='hour', aggfunc='mean')
sns.heatmap(pivot, cmap='Reds')
```

```python
# by state / region 
df.groupby('state')['arrival_delay_m'].mean().sort_values(ascending=False)
```

```python
# Category effects
df.groupby('category')['arrival_delay_m'].mean().sort_index()
```

```python
findings = {
    'overall_pct_delayed': df['is_delayed'].mean(),
    'avg_delay_minutes': df['arrival_delay_m'].mean(),
    'worst_station': station_stats.index[0],
    'worst_state': df.groupby('state')['arrival_delay_m'].mean().idxmax(),
    'worst_hour': df.groupby('hour')['arrival_delay_m'].mean().idxmax(),
    'worst_weekday': df.groupby('weekday')['arrival_delay_m'].mean().idxmax(),
    'worst_train_type': df.groupby('train_type')['arrival_delay_m'].mean().idxmax(),
}
findings
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
#overall delay distribution

fig, ax = plt.subplots()
sns.histplot(df['arrival_delay_m'].clip(-10, 60), bins=50, ax=ax)
ax.axvline(6, color='red', linestyle='--', label='DB punctuality threshold (6 min)')
ax.set_title('Distribution of Arrival Delays')
ax.set_xlabel('Delay (minutes)')
ax.legend()
plt.show()
```

```python
#top 15 worst station
top_stations = station_stats.head(15)

fig, ax = plt.subplots()
sns.barplot(x=top_stations['avg_delay'], y=top_stations.index, ax=ax, palette='Reds_r')
ax.set_title('Top 15 Stations by Average Delay')
ax.set_xlabel('Average Delay (minutes)')
plt.tight_layout()
plt.show()
```

```python
# Delay by hour of day (line chart)
hourly = df.groupby('hour')['arrival_delay_m'].mean()

fig, ax = plt.subplots()
hourly.plot(kind='line', marker='o', ax=ax, color='darkred')
ax.set_title('Average Delay by Hour of Day')
ax.set_xlabel('Hour')
ax.set_ylabel('Avg Delay (minutes)')
ax.set_xticks(range(0, 24))
plt.show()
```

```python
#delay by weekday
weekday_order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
weekday_avg = df.groupby('weekday')['arrival_delay_m'].mean().reindex(weekday_order)

fig, ax = plt.subplots()
sns.barplot(x=weekday_avg.index, y=weekday_avg.values, ax=ax, palette='Blues_d')
ax.set_title('Average Delay by Weekday')
ax.set_ylabel('Avg Delay (minutes)')
plt.xticks(rotation=45)
plt.show()
```

```python
#delay by state (region)
state_avg = df.groupby('state')['arrival_delay_m'].mean().sort_values(ascending=False)

fig, ax = plt.subplots()
sns.barplot(x=state_avg.values, y=state_avg.index, ax=ax, palette='Oranges_r')
ax.set_title('Average Delay by State')
ax.set_xlabel('Avg Delay (minutes)')
plt.tight_layout()
plt.show()
```

```python
# delay by station category
fig, ax = plt.subplots()
sns.boxplot(x='category', y='arrival_delay_m', data=df[df['arrival_delay_m'].between(-10, 60)], ax=ax)
ax.set_title('Delay Distribution by Station Category')
ax.set_xlabel('Station Category (1=major hub, 5=small)')
ax.set_ylabel('Delay (minutes)')
plt.show()
```

```python
 # delay by train type 
type_avg = df.groupby('train_type')['arrival_delay_m'].mean().sort_values(ascending=False)

fig, ax = plt.subplots()
sns.barplot(x=type_avg.index, y=type_avg.values, ax=ax, palette='Purples_d')
ax.set_title('Average Delay by Train Type')
ax.set_ylabel('Avg Delay (minutes)')
plt.show()
```

```python
# Geographic scatter   
fig, ax = plt.subplots(figsize=(8, 10))
scatter = ax.scatter(df['long'], df['lat'], c=df['arrival_delay_m'].clip(0, 30), cmap='Reds', s=10, alpha=0.5)
plt.colorbar(scatter, label='Delay (min)')
ax.set_title('Geographic Distribution of Delays')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
plt.show()
```
